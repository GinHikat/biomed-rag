```python
import pandas as pd
import os
```

```python
NOTEBOOK_DIR= os.getcwd()
BASE_DIR= os.path.join(NOTEBOOK_DIR, "..", "..")
data= os.path.join(BASE_DIR, "data", "external", "bioasq", "task_b", "bioasq.csv")
PROCESSED_DATA_DIR= os.path.join(BASE_DIR, "data","processed","bioasq")
```

```python
import ast

df["ideal_answer"] = df["ideal_answer"].apply(ast.literal_eval)
```

```python
print(type(df["ideal_answer"].iloc[0]))
```

```python
import json 

dataset= []

for _, r in df.iloc[:200].iterrows():
    id= str(r["id"])
    q= r["body"]
    a= r["ideal_answer"][0]

    dataset.append({
        "quest_id": id,
        "question": q,
        "ideal_answer": a 
})
    
os.makedirs(PROCESSED_DATA_DIR, exist_ok=True)

OUTPUT_FILE= os.path.join(PROCESSED_DATA_DIR,"bioasq.json")



with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    json.dump(dataset, f, indent=2)
```
